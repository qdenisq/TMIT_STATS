# TIMIT DataBase Statistics 

## Parse Database

```python
import timit_stats as tist
import os
import matplotlib.pyplot as plt

%pylab inline

trans_directory = "../USC-TIMIT/EMA/Data/M1/trans"

t_starts, t_ends, phonemes_list, _, _ = tist.parse_trans_corpus(trans_directory)
    
print "TIMIT loaded and parsed succesfully"
```

## Phonemes Presented in Database

```python
phonemes_unique = list(set(phonemes_list))
print "Phonemes set: {0}.".format(len(phonemes_unique))
print "Phonemes number of occurences in the whole corpus: {0}.".format(len(phonemes_list))
print "Phonemes list: {0}.".format(phonemes_unique)
```

## Phonemes aka Gestures Distribution

```python
# count phonemes
phonemes_unique = list(set(phonemes_list))
phonemes_counter = {}
for ph in phonemes_unique:
    phonemes_counter[ph] = phonemes_list.count(ph)
    
# plot pie chart 
plt.rcParams['patch.linewidth'] = 1 
plt.rcParams['patch.edgecolor'] = "white" 
cmap = plt.get_cmap('Set3')
fig1, ax1 = plt.subplots(figsize=(20, 10))
colors = [cmap(i) for i in np.linspace(0, 1, len(phonemes_counter))]
ax1.pie(phonemes_counter.values(), labels=phonemes_counter.keys(), autopct='%1.1f%%', colors=colors,
        startangle=90)
ax1.axis('equal')
fig1.suptitle("Gesture occurences", fontsize=14, fontweight='bold')
plt.show()
```

## Gesture Time Distribution
